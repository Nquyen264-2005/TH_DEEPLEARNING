Bài 3: Sinh viên tìm hiểu và viết chương trình sử dụng các hàm sau numpy.greater,
numpy.greater_equal, numpy.less, và numpy.less_equal để so sánh 2 mảng 1 chiều, 2 chiều.

```python
import numpy as np

print("--- So sánh mảng 1 chiều ---")
arr1_1d = np.array([1, 2, 3, 4, 5])
arr2_1d = np.array([5, 2, 8, 1, 3])

print("Mảng 1: ", arr1_1d)
print("Mảng 2: ", arr2_1d)
print("\nArr1 > Arr2:", np.greater(arr1_1d, arr2_1d))
print("Arr1 >= Arr2:", np.greater_equal(arr1_1d, arr2_1d))

print("Arr1 < Arr2:", np.less(arr1_1d, arr2_1d))

print("Arr1 <= Arr2:", np.less_equal(arr1_1d, arr2_1d))

print("\n--- So sánh mảng 2 chiều ---")
arr1_2d = np.array([[10, 20, 30], [40, 50, 60]])
arr2_2d = np.array([[5, 25, 20], [40, 45, 70]])

print("Mảng 1:\n", arr1_2d)
print("Mảng 2:\n", arr2_2d)
print("\nArr1 > Arr2:\n", np.greater(arr1_2d, arr2_2d))
print("Arr1 >= Arr2:\n", np.greater_equal(arr1_2d, arr2_2d))
print("Arr1 < Arr2:\n", np.less(arr1_2d, arr2_2d))
print("Arr1 <= Arr2:\n", np.less_equal(arr1_2d, arr2_2d))
```

Bài 20: Tìm hiểu phép toán numpy.unionld, numpy.insersectld (hội, giao) và viết chương trình minh họa 2 hàm này trên 2 Series x, y ngẫu nhiên có 20 phần tử

```python
import numpy as np
import pandas as pd
x_raw = np.random.randint(0, 30, size=20)
y_raw = np.random.randint(15, 45, size=20)

x = pd.Series(x_raw.tolist())
y = pd.Series(y_raw.tolist())

print("Series x (20 phần tử ngẫu nhiên):")
print(x.to_numpy())
print("\nSeries y (20 phần tử ngẫu nhiên):")
print(y.to_numpy())
union_xy = np.union1d(x, y)
print("\nHợp (Union) của x và y:")
print(union_xy)

intersect_xy = np.intersect1d(x, y)
print("\nGiao (Intersection) của x và y:")
print(intersect_xy)
```

Bài 21: Viết chương trình Pandas thêm series y vào series x theo chiều dọc (vertical), chiều ngang (horizontal) (Tìm hiểu hàm append và concat của pandas)

```python
import pandas as pd
series_x = pd.Series([10, 20, 30], name='X_values')
series_y = pd.Series([40, 50, 60], name='Y_values')

print("Series x:")
print(series_x)
print("\nSeries y:")
print(series_y)
series_vertical_concat = pd.concat([series_x, series_y])
print("\nThêm Series y vào Series x theo chiều dọc (vertical concat):")
print(series_vertical_concat)
series_horizontal_concat = pd.concat([series_x, series_y], axis=1)
print("\nThêm Series y vào Series x theo chiều ngang (horizontal concat):")
print(series_horizontal_concat)
```
